### a. Identify and address duplicate entries, missing values, and, if necessary, eliminate potential outliers
- Information about data set
- Summary of the data set

```python
data.describe() 
```

```python
#dropping duplicate values - checking if there are any duplicate rows and dropping if any
data=data.drop_duplicates()
data.shape
# 25 rows has been dropped because of duplication values
```

```python
#check for missing values, count them and print the sum for every column
data.isnull().sum() 
#conclusion :there are no null values in this dataset (no missing values)
```

```python
# Handle issues related to zero values in the dataset
# According to data.describe(), we can observe that only "BlastFurnaceSlag","FlyAsh", and "Superplasticizer" have zero values. 
# We are checking the rows conating zero values.
print(data[data['BlastFurnaceSlag']==0].shape[0])
print(data[data['FlyAsh']==0].shape[0])
print(data[data['Superplasticizer']==0].shape[0])
```

```python
# replace with median or mean
data['BlastFurnaceSlag']=data['BlastFurnaceSlag'].replace(0,data['BlastFurnaceSlag'].median())#skewed distribution (median)
data['FlyAsh']=data['FlyAsh'].replace(0,data['FlyAsh'].mean())#median is zero, replce with mean
data['Superplasticizer']=data['Superplasticizer'].replace(0,data['Superplasticizer'].median())#skewed distribution (median)
```

```python
# Handling potential outliers and rescale the data appropriately
# We use Quantile Transformer : This method transforms the features to follow a uniform or a normal distribution.
# Therefore, for a given feature, this transformation tends to spread out the most frequent values. 
# It also reduces the impact of (marginal) outliers: this is therefore a robust preprocessing scheme.

from sklearn.preprocessing import QuantileTransformer
x=data
quantile  = QuantileTransformer()
X = quantile.fit_transform(x)
df_new=quantile.transform(X)
df_new=pd.DataFrame(X)
df_new.columns =['Cement', 'BlastFurnaceSlag','FlyAsh','Water','Superplasticizer','CoarseAggregate','FineAggregate','Age','CompressiveStrength']
df_new.head()
```

```python
# According to the following boxplot, we can see that the outlier issue has been resolved. 
fig, axes = plt.subplots(1, 9, figsize=(16, 4))  

for i, col in enumerate(df_new):
    sns.boxplot(df_new[col], ax=axes[i])
    axes[i].set_title(col)  

plt.tight_layout()  
plt.show()  
```

Split dataset into train (70%) and test (30%) and build predictive models to determine the strength of concrete using the following techniques: K-Nearest Neighbor Regressor, Random Forest, and XGBoost (20 points)

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
import xgboost as xgb
from sklearn.metrics import mean_squared_error
```

```python
target_name='CompressiveStrength'
y= df_new[target_name]#given predictions - training data 
X=df_new.drop(target_name,axis=1)#dropping the Outcome column and keeping all other columns as X
```

```python
X.head() 
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X,y,test_size=0.3,random_state=0)#splitting data in 70% train, 30%test
```

```python
X_train.shape,y_train.shape
```

```python
X_test.shape,y_test.shape
```

```python
# K-Nearest Neighbor Regressor
knn_regressor = KNeighborsRegressor()
knn_regressor.fit(X_train, y_train)
knn_predictions = knn_regressor.predict(X_test)
knn_rmse = mean_squared_error(y_test, knn_predictions, squared=False)


# Random Forest Regressor
rf_regressor = RandomForestRegressor()
rf_regressor.fit(X_train, y_train)
rf_predictions = rf_regressor.predict(X_test)
rf_rmse = mean_squared_error(y_test, rf_predictions, squared=False)


# XGBoost Regressor
xgb_regressor = xgb.XGBRegressor()
xgb_regressor.fit(X_train, y_train)
xgb_predictions = xgb_regressor.predict(X_test)
xgb_rmse = mean_squared_error(y_test, xgb_predictions, squared=False)

```

```python
from sklearn.ensemble import VotingRegressor
from sklearn.metrics import mean_squared_error


#  Individual regressor models
KNN_REG = KNeighborsRegressor()
RF_REG = RandomForestRegressor()
XGB_REG = xgb.XGBRegressor()

# voting regressor ensemble
voting_regressor = VotingRegressor(estimators=[
    ('knn', KNN_REG),
    ('rf', RF_REG),
    ('xgb', XGB_REG)
])

# Train the voting regressor on the training data
voting_regressor.fit(X_train, y_train)

#  Evaluate the performance of the ensemble on the testing data
y_pred = voting_regressor.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score
import numpy as np

# Fit individual regressor models
KNN_REG.fit(X_train, y_train)
RF_REG.fit(X_train, y_train)
XGB_REG.fit(X_train, y_train)

# Calculate metrics for individual regressors
models = {
    'Voting Regressor': y_pred,
    'KNN Regressor': KNN_REG.predict(X_test),
    'Random Forest Regressor': RF_REG.predict(X_test),
    'XGBoost Regressor': XGB_REG.predict(X_test)
}


metrics = {
    'RMSE': lambda y_true, y_pred: np.sqrt(mean_squared_error(y_true, y_pred)),
    'MAE': mean_absolute_error,
    'MSE': mean_squared_error,
    'R2 Accuracy': r2_score
}
results = {}

for metric_name, metric_func in metrics.items():
    results[metric_name] = {}
    for model_name, y_pred in models.items():
        results[metric_name][model_name] = metric_func(y_test, y_pred)

# Print the results
print("Compare the results of voting regressor with individual classifiers:")
for metric_name, metric_values in results.items():
    print("\n{}:".format(metric_name))
    for model_name, value in metric_values.items():
        print("{}: {:.4f}".format(model_name, value))
```

### Conclusion:  based on the provided metrics, the XGBoost Regressor and then Voting Regressor consistently perform the best across all evaluation criteria, while the KNN Regressor performs the worst.
